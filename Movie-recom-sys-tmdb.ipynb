### DEFINE A FUNCTION WHICH CAN GIVE US THE NAME ONLY THAT MATTERS US IN GENRES AND WE CAN SEE THAT THIS lIST IS IN STING FORMAT AND WE HAVE TO CONVERT THAT STRING TO LIST FOR THIS WE GONNA USE ```ast.literal_eval()```


````python
import ast
````

````python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L    
````

````python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
````

````python
movies['keywords'][0] 
````

````python
## THE NAME VALUE ONLY MATTER FOR US OF THIS TOO SO AGAIN GONNA APPLY THE CONVERT FUNCTION
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
````

````python
movies['cast'][0]
````

````python
## DO THE SAME WITH THE CAST 
movies['cast'] = movies['cast'].apply(convert)
````

````python
## SEE A MOVIE HAS NUMEROUS OF CAST MEMBERS BUT THE TOP3 IS MORE IMPORTANT SO WE WILL DEFINE A FUNCTION FOR THIS
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
````

````python
movies.head()
````

````python
movies['crew'][0]
````

````python
## WE CAN HAVE A NUMEROUS OF CREW MEMBERS BUT THE DIRECTOR IS THE IMPORTANT ONE
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
        
movies['crew'] = movies['crew'].apply(fetch_director)
````

````python
movies.head()
````

````python
## REPLACE THE SPACE WITH ""
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ", ""))
    return L1
````

````python
## APPLY COLLAPSE TO EVERY COLUMN
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
````

````python
movies.head()
````

````python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
````

````python
movies.head()
````

````python
## NOW WE JUST MERGE ALL THE COLUMNS AS A TAG COLUMN EXCEPT MOVIE_ID AND TITLE AND THEN WE USE THIS COLUMN FOR COMPARING SIMILARITY
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
````

## STEMMING

- NOW IF WE CHECK THE VOCAB THEN WE CAN SEE THERE ARE WORDS LIKE ACTION & ACTIONS AND BOTH ARE CONSIDERED AS A DIFFERENT WORD SO FOR THIS WE HAVE TO DO ```STEMMING``` 

````python
import nltk
````

````python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
````

````python
## DEFINE A HELPER FUNCTION TO APPLY THE STEMMING
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)    
````

````python
## SEE THIS EXAMPLE
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron'
    )
````

````python
## APPLY ON TAG COLUMN
new_df['tag'] = new_df['tag'].apply(stem)
````

## VECTORIZATION

````python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
````

````python
vectors = cv.fit_transform(new_df['tag']).toarray()
````

## COSINE SIMILARITY

````python
from sklearn.metrics.pairwise import cosine_similarity
````

````python
similarity = cosine_similarity(vectors)
````

````python
similarity,similarity.shape
````

````python
## DEFINE A FUNCTION WHICH WILL RECOMMEND US THE FIVE MOVIE WHEN WE GIVE IT A MOVIE NAME
def recommender(movie_name):
    movie = movie_name.lower()
    movie_index = new_df[new_df['title'] == movie].index[0]
    if movie_index is not None:
        distance = similarity[movie_index]
        enumerate_list = list(enumerate(distance))
        recommender_list = sorted(enumerate_list,reverse=True,key=lambda x:x[1]) ## key element help us to sort it on the basis of similarity score not index
    
       
        for m_index,sim_score in recommender_list[1:6]:
            print(new_df.iloc[m_index].title.title())
    else:
        print("SORRY WE HAVE NO RECOMMENDATION FOR THIS ONE, HOPW YOU DON'T MIND!!🥺🥺🥺")
        
    
````
